# Savitsky Golay Filter Application

### Imports

```python
import numpy as np
import pandas as pd
from scipy import signal
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#model: Y = 3X + 4
#size of training data and scale of random noise 
pts=25 
noisescale=.05

x=np.linspace(-5,5,num=pts)

yactual_1 = np.sin(x * 0.5*np.pi)
yactual_2 = -(0.5 * (x-2))**2
yactual_3 = -(0.5* (x+2))**2
yactual = yactual_1 + yactual_2 + yactual_3
#np.random.seed(123) for reproducible 
#add noise scaled to 25% of range to data
yrand=yactual+noisescale*(yactual.max()-yactual.min())*np.random.normal(size=pts)

plt.plot(x, yrand, 'o')
plt.plot(x, yactual)
```

```python
ysavgol = signal.savgol_filter(yrand, 5, 3)
plt.plot(x, yrand, 'o')
plt.plot(x,ysavgol)
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
rmse(ysavgol, yrand)
```
